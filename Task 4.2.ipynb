Visualize the relationship between fundamental and overtones with a fundamental at 1Hz and adding the following four overtones to the graph.

```python
import numpy as np
import matplotlib.pyplot as plt

fundamental = 1
overtones = [fundamental * n for n in range(1, 5)]

x = np.linspace(0, 2  *np.pi, 200)

for n, frequency in enumerate(overtones):
    # draw overtones with different amplitudes to avoid noodle soup
    scale = (4 - n) / 4
    plt.plot(x, np.sin(frequency * x) * scale, label=f'{n+1}. harmonic')
plt.legend()
```
